**5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
size = 100
X1 = np.random.normal(0, 1, size)
X2 = 2 * X1 + np.random.normal(0, 0.1, size)
X3 = -1.5 * X1 + np.random.normal(0, 0.2, size)

data = pd.DataFrame({'Feature1': X1, 'Feature2': X2, 'Feature3': X3})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate the z data
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Generate a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot of z = sin(√(x² + y²))')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5, c=df['Population'], cmap='viridis')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Countries by Population and GDP')

# Annotate each point with the country name
for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=9, ha='right')

plt.colorbar(scatter, label='Population')
plt.show()
```

**1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.**

**2. .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.**


```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap

# Use output_notebook to render the plot inline in Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.rand(100) * 30 + 10
colors = np.random.rand(100) * 100

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot with a title and axis labels
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with sizes and colors
mapper = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))
p.scatter('x', 'y', size='sizes', color=mapper, alpha=0.6, source=source, legend_label='Random Data')

# Show the results
show(p)
```

**3.  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Use output_notebook to render the plot inline in Jupyter Notebook
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field='fruits',
       line_color='white', fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the results
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Use output_notebook to render the plot inline in Jupyter Notebook
output_notebook()

# Generate the data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Data Distribution", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad renderer with hist, edges
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white', alpha=0.7)

# Show the results
show(p)
```

**5. Create a Bokeh heatmap using the provided dataset.**

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar, ColumnDataSource
from bokeh.transform import transform

# Use output_notebook to render the plot inline in Jupyter Notebook
output_notebook()

# Generate the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a DataFrame for ColumnDataSource
data = {'x': x_flat, 'y': y_flat, 'z': z_flat}
source = ColumnDataSource(data)

# Create a new plot with a title and axis labels
p = figure(title="Heatmap of Data Distribution", x_axis_label='X', y_axis_label='Y',
           tools="hover", tooltips=[("Value", "@z")])

# Define a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=z_flat.min(), high=z_flat.max())

# Add a quad glyph for the heatmap
p.rect('x', 'y', width=0.1, height=0.1, source=source,
       fill_color=transform('z', mapper), line_color=None)

# Add a color bar to the right side of the plot
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
